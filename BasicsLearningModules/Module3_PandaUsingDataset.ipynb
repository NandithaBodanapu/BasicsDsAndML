```python
#Exploring Pandas using movie dataset 
import pandas as pd
import numpy as np


imdb_df=pd.read_csv("/Users/nandithabodanapu/PycharmJupyterProjects/BasicsDsAndML/data/IMDB-Movie-Data.csv")
# list the first four 
print("number of rows and columns ",imdb_df.shape)
print("the first 4 rows------")
print(imdb_df.head(5))# list the first 5 rows
print("Info about the dataset-----")
imdb_df.info()# provide info about the dataser, like number of rows data type etc
print("describing the dataset-------")
print(imdb_df.describe()) # provides basic statistical information
print("printing the last three rows-----")
imdb_df.tail(5)# list the last 5 records
```

```python
#Using the output of info , we can choose a particular column
print("printinh single column as a series ",imdb_df['Title'], )
print("printing more columns as series ",imdb_df['Genre'], imdb_df['Year'], imdb_df['Rating'])# lists one column after another
# listing multiple columns as data frame object will display it as a mini table
print("printing multiple columns as a dataframe object", imdb_df[['Genre','Year','Rating']])
```

```python
#selecting rows by using loc and iloc
print(imdb_df.loc[24])
print(imdb_df.iloc[24])
#we can set index,so the records will print based on index_value
rank_index_imdb_df=imdb_df.set_index('Rank')
print(rank_index_imdb_df.loc[24])
print(rank_index_imdb_df.iloc[24])

#to print multiple rows []
print(rank_index_imdb_df.loc[24:30])
```

```python
#using conditinal statements
print(imdb_df[((imdb_df["Year"]== 2006) & (imdb_df["Rank"]< 200))])
```

```python
#GROUPING AND SORTING
#Grouping split - apply-combine stategy
# so data divided into group, function is applied to each group,results are applied in new dataframe
#print(rank_index_imdb_df.groupby('Director').sum()) # bad call as it add years to which is not accurate

rank_index_imdb_df.groupby('Year')[["Rating"]].max()
```

```python
#SORTING
rank_index_imdb_df.sort_values(['Revenue (Millions)'], ascending=False)
```

```python
# checking for Null values
rank_index_imdb_df.isnull().sum().sort_values(ascending=False)
```

```python
metascore_mean=rank_index_imdb_df['Metascore'].mean()
rank_index_imdb_df['Metascore'].fillna(metascore_mean, inplace=True)
rank_index_imdb_df.isnull().sum().sort_values(ascending=False)
```
