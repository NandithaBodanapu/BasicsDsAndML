```python
#scatter plot
x=np.random.randn(100)
y=np.random.randn(100)
clr=np.random.randn(100)
s_z=(np.random.randn(100))*1000
#simple way
plt.scatter(x,y)
#adding more features
#plt.scatter(x,y,c=clr,s=s_z,alpha=0.8,cmap='viridis')
plt.style.use('classic')
plt.colorbar()
plt.show()
```

```python
#line plot
# y axis
student_1 =[80,89,75,90,67]
student_2=[78,60,79,90,89]
#x axis
subject=["math","science","social",'english','hindi']
#visualizing the data
fig,ax=plt.subplots()

ax.plot(subject,student_1,color='blue')
ax.plot(subject,student_2,color='red')
ax.set_ylim(55,95)
ax.grid(True)
plt.title("student marks comparison")
plt.xlabel("subject")
plt.ylabel("marks")
plt.style.use('ggplot')

plt.show()


```

```python
# histograms
hist_data=np.random.randn(100)
plt.hist(hist_data,bins=15,edgecolor='brown',alpha=0.2)
```

```python
#bar plot
#tv series
days=['Monday', 'Tuesday','Wednesday','Thursday','Friday']
hours_watched=[5,2,4,6,3]
#convert days to indexes
plt.title("hours of tv watched from monday to friday")
index_days=np.arange(len(days))
plt.bar(index_days,hours_watched)
plt.xticks(index_days,days)# this is to label the ticks with corresponding days
plt.show()
```

```python
#box plot
data1_ages=[10,50,65,20,45,35,53,46,78,23]
data2_ages=[20,40,15,25,40,75,23,36,88,13]
data3_ages=[80,70,55,30,25,15,53,86,58,43]
data_plot=[data1_ages,data2_ages,data3_ages]
plt.title("Age Box Plot")
plt.boxplot(data_plot)
plt.xlabel('groups')
plt.ylabel('Age number')

plt.show()
```
